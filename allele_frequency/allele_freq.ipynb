```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def plot_mutation_selection_dynamics(s=0.1, mu=0.01, nu=0.001, p0=0.5, t_max=100):
    def allele_dynamics(t, p, s, mu, nu):
        return s * p * (1 - p) - mu * p + nu * (1 - p)

    t_span = (0, t_max)
    t_eval = np.linspace(*t_span, 500)

    sol = solve_ivp(allele_dynamics, t_span, [p0], args=(s, mu, nu), t_eval=t_eval)

    plt.figure(figsize=(10, 5))
    plt.plot(sol.t, sol.y[0], label='Allele A Frequency (p)', color='blue')
    plt.xlabel("Time")
    plt.ylabel("Allele Frequency")
    plt.title("Allele Frequency Dynamics: Mutation-Selection Balance")
    plt.grid(True)
    plt.ylim(0, 1)
    plt.legend()
    plt.show()
```

```python
plot_mutation_selection_dynamics()
```

```python
plot_mutation_selection_dynamics(s=0.2, mu=0.005, nu=0.002, p0=0.2, t_max=200)
```
